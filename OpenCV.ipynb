## OpenCVの基礎
OpenCVは画像認識（Computer Vision）用のC/C++ライブラリ． 今回はPythonからその機能にアクセスする．

### イメージの読み込み

```python
img = cv2.imread("./dataset/standard_test_images/mandril_color.tif", cv2.IMREAD_UNCHANGED)
```

pythonから呼び出したOpenCVでは画像を表現するためにnumpyのarrayが利用されている．

```python
# imgの正体はnumpy.ndarray
img
```

### イメージの表示
OpenCVでは画素値はB（青），G（緑），R（赤）で格納されている．  
matplotlibでは画素値はRGBで格納されていると仮定しているので，画素値の順番を変換する必要がある．

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

### イメージの書き出し

```python
# OpenCVの関数を利用して書き出し
cv2.imwrite("./dataset/tmp/mandril_color_cv.tiff", img)
```

```python
# matplotlibの関数を利用して書き出し．RGBを仮定するのでオリジナルと異なる色で保存される．
plt.imsave("./dataset/tmp/mandril_color_plt_BGR.tiff",img)
```

```python
# matplotlibの関数を利用して書き出し．表示と同様に画素値をRGBに変更すると良い．
plt.imsave("./dataset/tmp/mandril_color_plt_RGB.tiff",cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

### イメージ情報の取得

```python
#  高さ，幅，チャネル数
height, width, channels = img.shape
print("height: "+str(height))
print("width: "+str(width))
print("channels: "+str(channels))

# 配列の要素数 = height * width * channels
print("size: "+str(img.size))
```

### ピクセル値の取得
特定の画素のBGR値をnumy.ndarray形式で取得できる．  

```python
pixVal = img[100,100]
pixVal
```

### ピクセル操作

```python
# ある画素を書き換える
img[100,100] = [0,0,0]
img[100,100]
```

```python
# ある範囲の画素値すべてを書き換える
img[50:150, 50:150] = [0,0,0]
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```
